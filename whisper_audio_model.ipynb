```python
from openai import OpenAI
client=OpenAI(api_key=openai.api_key)
```

```python
response=client.audio.speech.create(
    model="tts-1",
    voice="alloy",
    input="Babuu baccha hai tu mera "
)
response.stream_to_file("babbu.mp3")
```

```python
audio_file=open("/content/babbu.mp3","rb")
response=client.audio.transcriptions.create(
    model="whisper-1",
    file=audio_file,

)
print(response.text)
```
