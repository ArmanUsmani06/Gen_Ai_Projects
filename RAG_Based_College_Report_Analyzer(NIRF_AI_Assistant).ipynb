```python
base_instruction='you are a helpful assistant which help in answering some questions from a document i provided as input, these questions are related to my college nirf report .you have to read the full docx and answer my question from the document only . if any information which is asked is not present kindly say "i dont know i dont give any random information" only and nothing else'
```

```python
question='the document shared is of which year?'
```

```python
question='number of students enrolled in session 2025-2026? give me detialed breakdown.'
```

```python
question='is there a % increase in no. of students enrolled in consecutive year or not. give the complete comprehensive calculation'
```

```python
question='give me the complete faculty details who has the designation as professor'
```

```python
question='number of total btech,mtech,phd holders students across all years'
```

```python
question='give me the detail breakdown of fee details for btech '
```

```python
prompt=base_instruction+"\n\n"+"question:{0}".format(question)+"\n\n"+"transcript\n:{0}".format(content)
```

```python
print(prompt)
```

```python
message=[{"role":"user","content":prompt}]
```

```python
chat_response = openai.chat.completions.create(
    model="gpt-3.5-turbo-16k",      # or another chat model you have
    messages=message,
    max_tokens=1000,
    temperature=0.5,
    n=1,
    stop=None,
    frequency_penalty=0,
    presence_penalty=0
)

print(chat_response.choices[0].message.content)
```
